```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"MRT_201909.csv",delimiter=r"\s+")
```

```python
data.columns = ["Date","Hour","From","To","Number"]
```

```python
i1 = (data["From"] == "新店") & (data["To"] == "台北車站")
```

```python
data_i1=data[i1]
```

```python
a_num1 = []
x1=[]
for i in range(24):
    avg1 = np.mean(data_i1[(data_i1.Hour ==i)].Number)
    if np.isnan(avg1):
        avg1 = 0
    a_num1.append(avg1)
    x1.append(i)
```

```python
plt.plot(x1,a_num1)
plt.xticks(range(24))
plt.show()
```
