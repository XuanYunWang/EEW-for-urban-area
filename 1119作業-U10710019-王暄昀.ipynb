```python
import matplotlib.pyplot as plt
from matplotlib.transforms import blended_transform_factory
from obspy import Stream
from obspy import read
from obspy.geodetics import gps2dist_azimuth
import os
%matplotlib inline
```

```python
ddir ="D:\\python\\1119\\data"
files=os.listdir(ddir)
```

```python
ff = []
for i in files:
    nn = i.find("BHZ")
    if nn > 0:
        print (i)
        ff.append(ddir+"\\"+i)
```

```python
eq_lat = -4.9521
eq_lon = 94.3299
ev_coord1 = (eq_lat,eq_lon)
```

```python
st=Stream()
for i in ff:
    st += read(i)
```

```python
for tr in st:
    tr.stats.distance = gps2dist_azimuth(tr.stats.sac.stla, tr.stats.sac.stlo,eq_lat, eq_lon)[0]
```

```python
st.plot(type='section', plot_dx = 2000*1000, recordlength=2000,time_down=False, linewidth=.25, grid_linewidth=.25,scale=3.5)
```
