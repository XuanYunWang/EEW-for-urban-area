```python
import obspy
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
AA=glob.glob("d:\\python\\1217\\*.SAC")
```

```python
AA
```

```python
st_all = obspy.Stream()
for i in AA:
    st_all = st_all+obspy.read(i)
```

```python
st_all[0].filter('lowpass', freq=1.0, corners=2, zerophase=True)
st_all[0].plot()
```

```python
fig = plt.figure(figsize=(15,15), dpi=100) 
num = 1
for i in st_all[0:3]:

    y=i.data
    x=i.times()
    
    plt.subplot(3,1,num)
    plt.plot(x,y, color="c")
    
    num = num + 1
plt.savefig("1217-U10710019-.png")
```
