```python
#週期3個單位
import matplotlib.pyplot as plt
import numpy as np


t=np.linspace(0,20,200)
y=np.cos(2/3*np.pi*t)
    
plt.plot(t,y,color="g")
plt.title("Period=3")
plt.xlabel("time")
plt.ylabel("amplitude")
```

```python
#波長11個單位
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(0,20,200)
y=np.cos(2/11*np.pi*x)
    
plt.plot(x,y,color="c")
plt.title("wavelength=11")
plt.xlabel("distance")
plt.ylabel("amplitude")
```

```python

import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(0,20,200)
y=np.cos(2/11*np.pi*x)

x1=np.linspace(0,20,200)
y1=np.cos(2/11*np.pi*x-2*np.pi/3)

x2=np.linspace(0,20,200)
y2=np.cos(2/11*np.pi*x-4*np.pi/3)

plt.plot(x,y,color="c",label='t=0')
plt.plot(x1,y1,color="g",label='t=1')
plt.plot(x2,y2,color="y",label='t=2')

plt.title("wavelength=11,Period=3")
plt.xlabel("distance")
plt.ylabel("amplitude")
plt.legend(loc='upper right')
plt.show()
```
